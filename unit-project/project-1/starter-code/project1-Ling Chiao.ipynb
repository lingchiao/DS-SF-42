#### 3a. How might outliers impact your analysis? 

Answer: 
- Outliers will make the mean not a representative average for your dataset. If there are low outliers in GPV, it will drag down the average
- When looking for correlation between 2 variables, outliers that don't follow the same pattern as the rest of the dataset will make the correlation/colinearity unreliable. Lets say 2 metrics are completely correlated (colinearity = 1), but there is 1 outlier, the correlation without removing the outlier will not show the 2 variables are not as correlated

#### 3b. How will you test for outliers? 

Answer:

- Using the 1.5*IQR rule. If a data point is more than 1.5 * IQR above the third quartile or 1.5 * IQR below the first quartile, then it is considered an outlier
- Box and whisker plot is helpful here. It will easily show outliers as dots that are separate from the rest of the data

#### 4a. What is colinearity? 

Answer: 
- Colinearity is the correlation of 2 different variables. If 2 variables are perfectly correlated (ex: pounds vs. kg), then the colinearity will be 1 or -1. This means 1 predictor variable can be exactly predicted from the other with 100% accuracy (in this case if we had kgs, we know that pounds is exactly 1kg * 2.20462)

#### 4b. How will you test for colinearity? 

Answer:

- We can use the .corr() function to check for colinearity between 2 variables
- Plot data in a scatter graph to see if the dots follow a pattern of a line

#### Risks and and Assumptions of Data

- We are assuming that these variables (GRE, GPA and alma mater prestige) are the ones that will imapct admission. If the admission board actually do not care about these variables at all, and instead care about age, extra-cirricular activities, volunteer hours, then this will not be a good model. Even if we do find a correlation, it could just be random and the model will still not be a good model to predict admission probability.
- We are assuming that this data is a good representation of the total population. If that is not true or if there is missing data, the model that we come up with will not be a good model to predict student admission for *all* students
- If colinearity is quite low between the predictors and the value we are trying to predict. Even if we did come up with a model it may not be a very accurate model at predicting if a student was admitted or not
